```python
from urllib.request import urlopen
import json
from pandas import json_normalize
import pandas as pd
```

Register of inhabitants. Population of the city of Barcelona by sex and academic level

```python
url1 = 'https://opendata-ajuntament.barcelona.cat/data/api/action/datastore_search?resource_id=ac73b03b-247c-4c05-b773-72b3ca25dc74&limit=876'
fileobj1 = json.load(urlopen(url1))
#print(fileobj1)
```

```python
fileobj1.keys()
```

```python
fileobj1['result'].keys()
```

```python
sub_df1 = pd.json_normalize(fileobj1['result']['records'])
```

```python
sub_df1
```

```python
sub_df1['Nivell_academic'].value_counts()
```

Elementary school / school graduate / ESO / FPI 

Without studies 

High school / BUP / COU / FPII / CFGM Intermediate Grade

No records 

University studies / CFGS higher degree

Primary education / certificate of schooling / EGB

```python
sub_df1 = sub_df1.rename(columns = {"Nom_Districte": "District", 
                                    "Codi_Districte" : "District_Code", 
                                    "Nom_Barri": "Neighborhood",
                                    "Codi_Barri": "Neighborhood_Code",
                                    "Sexe": "Sex",
                                    "Nivell_academic": "Academic_Qualification",
                                    "Nombre":"Population"})
```

```python
educational_stages = {"Batxillerat elemental / graduat escolar / ESO / FPI": "Elementary school / school graduate / ESO / FPI",
       "Sense estudis": "Without studies",
       "No consta": "No records",
       "Batxillerat superior / BUP / COU / FPII / CFGM grau mitjà": "High school / BUP / COU / FPII / CFGM Intermediate Grade",
       "Estudis primaris / certificat d'escolaritat / EGB": "Primary education / certificate of schooling / EGB",
       "Estudis universitaris / CFGS grau superior": "University studies / CFGS higher degree"}
```

```python
sub_df1["Academic_Qualification"] = sub_df1["Academic_Qualification"].map(educational_stages)
```

```python
sex = {"Dona": "Women",
       "Home": "Men"}
```

```python
sub_df1["Sex"] = sub_df1["Sex"].map(sex)
```

```python
table1 = sub_df1.drop(columns=["_id", "Any"])
```

```python
table1
```

```python
#table1.groupby(['District', 'District_Code', 'Neighborhood', 'Neighborhood_Code', 'Sex', 'Academic_Qualification', 'Population']).sum()
```
